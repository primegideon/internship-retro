```python
# 2. Import libraries and set up the BigQuery client
from google.cloud import bigquery
import pandas as pd

# Your specific Google Cloud Project ID
project_id = 'ride-share-demand-analysis'
client = bigquery.Client(project=project_id)

# 3. Write the safe, filtered SQL query
sql_query = """
SELECT
trip_start_timestamp,
    fare,
    tips,
    trip_total,
    trip_seconds,
    pickup_community_area
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
WHERE EXTRACT(YEAR FROM trip_start_timestamp) = 2023
  AND EXTRACT(MONTH FROM trip_start_timestamp) = 1
LIMIT 1000;
"""

# 4. Run the query and load the results into a pandas DataFrame
print("Fetching data from BigQuery...")
df = client.query(sql_query).to_dataframe()

# 5. Display the first 5 rows to confirm it worked!
print("Data loaded successfully! Here is a preview:")
df.head()
```
